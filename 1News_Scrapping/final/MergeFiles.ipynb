```python
from pathlib import Path
import pandas as pd
import numpy as np
import os

path = os.getcwd()

# Get the files from the path provided in the OP
files = Path(path).glob('*.csv') 
```

```python
import pandas as pd
dfs = list()
for i, f in enumerate(files):
    data = pd.read_csv(f)
    dfs.append(data)

df = pd.concat(dfs, ignore_index=True)
```
